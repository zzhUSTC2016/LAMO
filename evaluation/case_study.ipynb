```python
import json
from matplotlib import pyplot as plt

data_path = '../data/finetune_data_ratio_2_ips.json'
with open(data_path, 'r') as f:
    data = json.load(f)

print('Number of data points:', len(data))
```

```python
sample = {'input': "\nAge: 48,\nGender: female,\nDiagnose: ['Pericardial disease NOS', 'Pleural effusion NOS', 'Cardiogenic shock', 'Hypotension NOS', 'Depressive disorder NEC', 'Cervical disc displacmnt', 'Joint pain-l/leg', 'Hyperlipidemia NEC/NOS'],\nProcedure: ['Venous cath NEC', 'Dx ultrasound-heart', 'Pericardiectomy'].\nCandidate drug: Fluticasone propionate.\n",
 'output': 'No.'}

# outputYesNo
pos_med_dict = {}
neg_med_dict = {}

for sample in data:
    candicate_drug = sample['input'].split('\n')[-2].split(':')[-1].strip('\n .')
    output = sample['output'].strip('\n .')
    if output == 'Yes':
        if candicate_drug in pos_med_dict:
            pos_med_dict[candicate_drug] += 1
        else:
            pos_med_dict[candicate_drug] = 1
    elif output == 'No':
        if candicate_drug in neg_med_dict:
            neg_med_dict[candicate_drug] += 1
        else:
            neg_med_dict[candicate_drug] = 1

# sort pos_med_dict by the med count
pos_med_dict = {k: v for k, v in sorted(
            pos_med_dict.items(), key=lambda item: item[1], reverse=True)}
neg_med_dict = {k: neg_med_dict[k] for k in pos_med_dict.keys()}
# plt.figure(figsize=(20, 10))
x = range(len(pos_med_dict))
plt.plot(x, pos_med_dict.values(), color='g', label = 'Yes')
plt.plot(x, neg_med_dict.values(), color='r', label = 'No')
plt.legend()
```
